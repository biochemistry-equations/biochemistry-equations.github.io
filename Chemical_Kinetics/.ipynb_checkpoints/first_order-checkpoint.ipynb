#### first order reaction 

The rate law is:

$$
\frac{d[A]}{dt} = -k[A]
$$

Integrating gives:

$$
\ln[A] = \ln[A]_0 - kt
$$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 10)
A0 = 1.0
k = 0.3
A = A0 * np.exp(-k * t)

plt.figure(figsize = (2,2))
plt.plot(t, A, '-ok')
plt.xlabel("Time")
plt.ylabel("[A]")
plt.title("First-Order Decay")
plt.show()
```
